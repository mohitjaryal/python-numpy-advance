#### Checking the speed of NumPy array and Python Lists

```python
import time
import numpy as np
```

```python
# checking the speed of python lists

# creating python lists
a = [i for i in range(10000000)]
b =[i for i in range(10000000,20000000)]

c =  []


start = time.time()

for i in range(len(a)):
    c.append(a[i] + b[i])
print(time.time() - start)
```

```python
# checking the speed of numpy array

arr1 = np.arange(10000000)
arr2 = np.arange(10000000,20000000)

c =  []


start = time.time()
c = arr1 + arr2
print(time.time() - start)
```

#### Now checking memory for the python lists and numpy arrays

```python
# lists

li = [i for i in range(10000000)]


# sys means system
import sys

# checking memory occuped by the system
sys.getsizeof(li)
```

```python
# numpy array

arr = np.arange(10000000)

# checking the memory occupied by the code
sys.getsizeof(arr)


# we can optimise this more by changing the int type from 64 bit to 32 bit
# numpy array

arr = np.arange(10000000,dtype=np.int32)

# checking the memory occupied by the code
sys.getsizeof(arr)
```

### Advance Indexing

#### Fancy indexing

```python
# creating array
a = np.arange(24).reshape(6,4)
a
```

```python
# print 1,3,4 row from the above array

# arrayname[[index]]
a[[0,2,3]]
```

```python
# print 1,3,4,6 row
a[[0,2,3,5]]
```

```python
# print 1,3,4 column
a[:,[0,2,3]]
```

#### Bollean Indexing

```python
# creating array

# syntax of randint -> .randint ( start(range), end(range), how many items we want)
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
# find all numbers greater than 50
b[b > 50]
```

```python
# find all the even numbers
b[b % 2 == 0]
```

```python
# find all numbers greater than 50 and are even
b[(b > 50) & (b % 2 ==0)]
```

```python
# find all numbers greate than 50 and are odd
b[(b>50) & (b % 2 !=0)]
```

```python
# find all numbers divisible by 7
b[b % 7 == 0]
```

```python
# find all numbers not divisible by 7
b[~(b % 7 == 0)]
```

#### Broadcasting

#### It describes how NumPy treats arrayof different shapes during arithmetic operations

```python
# creating arrays
a = np.arange(12).reshape(4,3)
b = np.arange(3)


print(a+b)
```

```python
# this will not work because shapes are not same
a = np.arange(12).reshape(4,3)
b = np.arange(4)
```

```python
# another example

# this will work

a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a+b)
```

```python
# another example

# this will work
a = np.array([1]) # shape (1,1)
b = np.arange(4).reshape(2,2) # shape (2,2)

print(a+b)
```

#### Checking the speed of NumPy array and Python Lists

```python
# using time module
import time
import numpy as np

# checking the speed of python lists

# creating python lists
a = [i for i in range(10000000)]
b =[i for i in range(10000000,20000000)]

c =  []


start = time.time()

for i in range(len(a)):
    c.append(a[i] + b[i])
print(time.time() - start)
```

```python
# checking the speed of numpy array

arr1 = np.arange(10000000)
arr2 = np.arange(10000000,20000000)

c =  []


start = time.time()
c = arr1 + arr2
print(time.time() - start)
```

#### Now checking memory for the python lists and numpy arrays

```python
# lists

li = [i for i in range(10000000)]


# sys means system
import sys

# checking memory occuped by the system
sys.getsizeof(li)
```

```python
# numpy array

arr = np.arange(10000000)

# checking the memory occupied by the code
sys.getsizeof(arr)


# we can optimise this more by changing the int type from 64 bit to 32 bit
# numpy array

arr = np.arange(10000000,dtype=np.int32)

# checking the memory occupied by the code
sys.getsizeof(arr)
```
